```python
import kagglehub
import pandas as pd
import numpy as np
import os

path = kagglehub.dataset_download("fredericobreno/play-tennis")
print("Path to dataset directory:", path)

csv_path = os.path.join(path, "play_tennis.csv")

# Load the data
data = pd.read_csv(csv_path)
data.head()
```

```python
data
```

```python
# Problem 1
# outlook = sunny , Temp = Hot, Humidity = High, Wind = Weak
# play or not?
```

```python
# solution
# P(Yes|Sunny,Hot,High,Weak) = P(Yes) * P(Sunny|Yes) * P(Hot|Yes) * P(High|Yes) * P(Weak|Yes)
# P(NO|Sunny,Hot,High,Weak) = P(NO) * P(Sunny|NO) * P(Hot|NO) * P(High|NO) * P(Weak|NO)

# Then We Will compare and decide using the maximum a posteriori rule
```

```python
# Problem 2
# outlook = overcast , Temp = cold, Humidity = low, Wind = Weak
# play or not?
```

```python
# solution
# P(Yes|Overcast, Cold, Low, Weak) = P(yes) * P(overcast|yes) * P(cold|Yes) * P(Low|Yes) * P(Weak|Yes)
# P(No|Overcast, Cold, Low, Weak) = P(No) * P(overcast|No) * P(cold|No) * P(Low|No) * P(Weak|No)
# Then We Will compare and decide using the maximum a posteriori rule
```

```python
#Every algorithm works on a training step and a testing step.
#In the training step, the Naive Bayes self-organizes a look-up table in which it stores all the probabilities in a dictionary.
```

```python
# P(Yes)
Prob_Of_Yes=data['play'].value_counts()['Yes']/len(data)
print(f"P(Yes): {Prob_Of_Yes}")
```

```python
# P(no)
Prob_of_NO=data['play'].value_counts()['No'] / len(data)
print(f"P(No): {Prob_of_NO}")
```

```python
# outlook
data['outlook'].unique()
```

```python
pd.crosstab(data['outlook'],data['play'])
```

```python
Prob_Overcast_Yes=data[data['play']=='Yes']['outlook'].value_counts()['Overcast']/data['play'].value_counts()['Yes']
Prob_Overcast_NO=data[data['play']=='No']['outlook'].value_counts().get('Overcast', 0)/data['play'].value_counts()['No']
print(Prob_Overcast_Yes,Prob_Overcast_NO,sep=" , ")
```

```python
Prob_Rain_Yes=data[data['play']=='Yes']['outlook'].value_counts()['Rain']/data['play'].value_counts()['Yes']
Prob_Rain_No=data[data['play']=='No']['outlook'].value_counts()['Rain']/data['play'].value_counts()['No']
print(f"P(Rain|Yes): {Prob_Rain_Yes} , P(Rain|No): {Prob_Rain_No} ")
```

```python
Prob_Sunny_Yes=data[data['play']=='Yes']['outlook'].value_counts()['Sunny']/data['play'].value_counts()['Yes']
Prob_Sunny_No=data[data['play']=='No']['outlook'].value_counts()['Sunny']/data['play'].value_counts()['No']
print(f"P(Sunny|Yes):{Prob_Sunny_Yes}, P(Sunny|No): {Prob_Sunny_No}")
```

```python
# Temperature
data['temp'].unique()
```

```python
pd.crosstab(data['temp'],data['play'])
```

```python
Prob_Cool_Yes=data[data['play']=="Yes"]['temp'].value_counts()['Cool']/data['play'].value_counts()['Yes']
Prob_Cool_No=data[data['play']=="No"]['temp'].value_counts()['Cool']/data['play'].value_counts()['No']
print(f"P(Cool|Yes): {Prob_Cool_Yes}, P(Cool|No): {Prob_Cool_No}")
```

```python
Prob_Hot_Yes=data[data['play']=="Yes"]['temp'].value_counts()['Hot']/data['play'].value_counts()['Yes']
Prob_Hot_No=data[data['play']=="No"]['temp'].value_counts()['Hot']/data['play'].value_counts()['No']
print(f"P(Hot|Yes): {Prob_Hot_Yes}, P(Hot|No): {Prob_Hot_No}")
```

```python
Prob_Mild_Yes=data[data['play']=="Yes"]['temp'].value_counts()['Mild']/data['play'].value_counts()['Yes']
Prob_Mild_No=data[data['play']=="No"]['temp'].value_counts()['Mild']/data['play'].value_counts()['No']
print(f"P(Mild|Yes): {Prob_Mild_Yes}, P(Mild|No): {Prob_Mild_No}")
```

```python
Prob_High_Yes=data[data['play']=="Yes"]['humidity'].value_counts()['High']/data['play'].value_counts()['Yes']
Prob_High_No=data[data['play']=="No"]['humidity'].value_counts()['High']/data['play'].value_counts()['Yes']
print(f"P(High|Yes): {Prob_High_Yes}, P(High|No): {Prob_High_No}")
```

```python
Prob_Normal_Yes=data[data['play']=="Yes"]['humidity'].value_counts()['Normal']/data['play'].value_counts()['Yes']
Prob_Normal_No=data[data['play']=="No"]['humidity'].value_counts()['Normal']/data['play'].value_counts()['No']
print(f"P(Normal|Yes): {Prob_Normal_Yes}, P(Normal|No): {Prob_Normal_No}")
```

```python
 # Wind
data['wind'].unique()
```

```python
Prob_Weak_Yes=data[data['play']=="Yes"]['wind'].value_counts()['Weak']/data['play'].value_counts()['Yes']
Prob_Weak_No=data[data['play']=="No"]['wind'].value_counts()['Weak']/data['play'].value_counts()['No']
print(f"P(Weak|Yes): {Prob_Weak_Yes}, P(Weak|No): {Prob_Weak_No}")
```

```python
Prob_Strong_Yes=data[data['play']=="Yes"]['wind'].value_counts()['Strong']/data['play'].value_counts()['Yes']
Prob_Strong_No=data[data['play']=="No"]['wind'].value_counts()['Strong']/data['play'].value_counts()['No']
print(f"P(Strong|Yes): {Prob_Strong_Yes}, P(Strong|No): {Prob_Strong_No}")
```

Solution

```python
# Problem 1
# outlook = sunny , Temp = Hot, Humidity = High, Wind = Weak
Prob_Yes_Sunny_Hot_High_Weak=Prob_Of_Yes * Prob_Sunny_Yes * Prob_Hot_Yes * Prob_High_Yes * Prob_Weak_Yes
print(f"Probability of Yes, when outlook = sunny , Temp = Hot, Humidity = High, Wind = Weak")
print("OR")
print(f"P(Yes|Sunny,Hot,Humidity,Weak): {Prob_Yes_Sunny_Hot_High_Weak}")
```

```python
Prob_No_Sunny_Hot_High_Weak = Prob_of_NO * Prob_Sunny_No * Prob_Hot_No * Prob_High_No * Prob_Weak_No
print(f"Probability of No, when outlook = sunny , Temp = Hot, Humidity = High, Wind = Weak")
print("OR")
print(f"P(No|Sunny,Hot,Humidity,Weak): {Prob_No_Sunny_Hot_High_Weak}")
```

```python
if(Prob_Yes_Sunny_Hot_High_Weak>Prob_No_Sunny_Hot_High_Weak):
    print("Yes")
else:
    print("No")
```

```python
# Problem 2
# outlook = overcast , Temp = cold, Humidity = Normal, Wind = Weak
# play or not?

Prob_Yes_Overcast_Cold_Normal_Weak=Prob_Of_Yes * Prob_Overcast_Yes * Prob_Cool_Yes * Prob_Normal_Yes * Prob_Weak_Yes
print(f"Probability of Yes, when outlook = overcast , Temp = cold, Humidity = Normal, Wind = Weak")
print("OR")
print(f"P(Yes|Overcast,Cold,Normal,Weak): {Prob_Yes_Overcast_Cold_Normal_Weak}")
```

```python
Prob_Yes_Overcast_Cold_Normal_Weak = Prob_of_NO * Prob_Overcast_NO * Prob_Cool_No * Prob_Normal_No * Prob_Weak_No
print(f"Probability of No, when outlook = overcast , Temp = cold, Humidity = Normal, Wind = Weak")
print("OR")
print(f"P(No|Overcast,Cold,Normal,Weak): {Prob_Yes_Overcast_Cold_Normal_Weak}")
```
